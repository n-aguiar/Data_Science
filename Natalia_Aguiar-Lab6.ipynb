```python
import numpy as np #importing libraries
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,50) #numbers we wanted to graph
```

```python
noise = 10 * np.random.randn(50,) #random numbers to make 
```

```python
y = 3 * x + 7 + noise #the equation
```

```python
plt.plot(x,y,'.'); #plotting the equation with noise
```

```python
m_guess = 1 #best fit line test
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
m_guess = 5 #best fit line test
b_guess = 15

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
m_guess = 3 #best fit line guess
b_guess = 7

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```

```python
loss = np.sum((y_guess - y)**2) #the correct loss
```

```python
for m_guess in np.linspace(-10,10,10): #guesses the best fit line without knowing the best value
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess) #prints the best fit line (values) with least errors
```

```python
best_loss = 100000 #best fit line test again 
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess) #best fit line out of all of them
```
